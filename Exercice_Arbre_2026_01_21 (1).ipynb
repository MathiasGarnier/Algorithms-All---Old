```python
# À CORRIGER / COMPLÉTER

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Conseil : Décommentez et vérifiez l'ordre des imports ou corrigez ce qu'il faut.
```

```python
# TODO : Charger le dataset Iris via load_iris
# Examinez son contenu et stocker les features dans X et la cible dans y

iris_data = load_iris()

print(iris_data.keys())

X = iris_data['data']
y = iris_data['target']
```

```python
print("Nom des features :", iris_data.feature_names)
print("Nom des classes :", iris_data.target_names)
print("Taille de X :", X.shape)
print("Taille de y :", y.shape)
```

```python
 # TODO : Corrigez l'appel à train_test_split pour avoir un split 70% (train) vs 30% (test)
#        et fixez un random_state (par exemple 42) pour la reproductibilité.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)

print("Taille X_train :", X_train.shape)
print("Taille X_test :", X_test.shape)
```

```python
# TODO : Complétez les arguments du DecisionTreeClassifier si nécessaire (profondeur suggérée : 3)
clf = DecisionTreeClassifier(
    criterion="entropy",
     max_depth=3,
     random_state=3)


# TODO : Entraînez le modèle sur X_train et y_train
clf.fit(X_train, y_train)
```

```python
# TODO : Faites la prédiction sur les données tests
y_pred = clf.predict(X_test)
```

```python
# TODO : Complétez l'appel à plot_tree
# utilisez les données incluses dans iris_data aidez-vous avec : print(iris_data.keys())
plt.figure(figsize=(12, 8))
plot_tree(
    clf,
    feature_names=iris_data.feature_names,  # Compléter
    class_names=iris_data.target_names,    # Compléter
    filled=True
)

plt.show()
```

```python
# comparez y_test et y_pred
import pandas as pd

df_comp = pd.DataFrame({
    "y_test": y_test,
    "y_pred": y_pred
})

df_comp
```

```python
# Matrice de confusion
cm = confusion_matrix(y_test, y_pred)
print("Matrice de confusion :\n", cm)

```

```python
# Calcul de l'accuracy
acc = accuracy_score(y_test, y_pred) # Compléter en vous inspirant de confusion_matrix
print(f"Accuracy : {acc:.2f}")

# Classification report
report = classification_report(y_test, y_pred, target_names=iris_data.target_names)
print("Classification Report :\n", report)
```
